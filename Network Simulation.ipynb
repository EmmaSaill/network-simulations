# Interbank Network Simulation

## Banks

```python
class Bank(object):
    def __init__(self, bank_id, interbank_assets=0.2, liabilities=0.96, default=1):
        self.bank_id = bank_id
        self.interbank_assets = interbank_assets
        self.liabilities = liabilities
        self.default = default # 0 = default, 1 = no default
        
    def __str__(self):
        return str(self.bank_id)
    
external_assets = 0.8
```

## Network

```python
from itertools import combinations
from networkx import nx
import matplotlib.pyplot as plt
import random

# probability for an edge to exist
p = 0.5

# number of nodes
n = 5
nodes = []
for i in range(n):
    newBank = Bank(i)
    nodes.append(newBank)

g = nx.Graph()
g.add_nodes_from(nodes)

for i, j in combinations(g, 2):
    if random.random() < p:
        g.add_edge(i, j, weight=0.3)
        print(i, j)

nx.draw(g, with_labels=True)
plt.show()
```

```python
def random_default():
    defaulting_bank = random.choice(nodes)
    print("Bank to default has Bank ID: ", defaulting_bank)

random_default()
```
